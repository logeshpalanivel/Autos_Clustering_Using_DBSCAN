```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.columns:
  if df[column].dtype == 'object':
    df[column]=le.fit_transform(df[column])
```

```python
df
```

```python
X = df.iloc[:, [0,1]].values
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps =1,min_samples=2,metric='euclidean')
```

```python
model = db.fit_predict(X)
```

```python
model
```

```python
labels = db.labels_
labels
```

```python
n = len(set(labels)) - (1 if -1 in labels else 0)
n
```

```python
plt.figure(figsize=(16,8))
plt.scatter(X[labels == 0,0], X[labels == 0,1],color = 'red' , label = 'cluster 1', s= 50)
plt.scatter(X[labels == 1,0],X[labels== 1,1],color = 'violet',label = 'cluster 2',s=50)
plt.scatter(X[labels == 2,0],X[labels== 2,1],color = 'black',label = 'cluster 3',s=50)
plt.scatter(X[labels == 3,0],X[labels== 3,1],color = 'blue',label = 'cluster 4',s=50)
plt.scatter(X[labels == 4,0],X[labels== 4,1],color = 'orange',label = 'cluster 5',s=50)
plt.xlabel('symboling')
plt.ylabel('make')
plt.title('DBSCAN')
plt.grid(False)
plt.legend()
plt.show()
```
